```python
# T test for gender
from scipy.stats import stats

genders = df['gender'].unique()
grouped_gender = df.groupby('gender')
for value in ['math score','reading score','writing score']:
    print('----------')
    print('comparison by {}'.format(value))
    print('----------')
    for i in range(0,len(genders)):
        for j in range(i+1,len(genders)):
            print('ttest between {} and {}:'.format(genders[i],genders[j]))
            print(stats.ttest_ind(
                df[df['gender']==genders[i]][value], 
                df[df['gender']==genders[j]][value]
            ))
    
```
