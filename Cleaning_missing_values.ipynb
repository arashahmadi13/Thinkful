# Working on Part (2)

```python
df1 = df.copy()
columns = df.columns
cleaning_list = columns[3:]
for column in cleaning_list:
    for value in column:
        df1[column] = df1[column].fillna(df1[column].mean())
df1.head()
```

# Working on part (3)

```python
df2 = df.copy()
years = df2['YEAR'].unique()
for colummn in cleaning_list:
    for year in years:
        df2.loc[df2["YEAR"] == year, column] = df2.loc[df2["YEAR"] == year, column].fillna(df2[df2["YEAR"] == year][column].mean())
```

# Working on part (4)

```python
df3 = df.copy()

for column in cleaning_list:
    df3[column] = df3[column].interpolate()
df3.head()
```

# Part (5)

```python
#Drawing all four df and compare them

fig = plt.figure(figsize=(20,100))

for value, column in enumerate(cleaning_list):
    df = pd.concat([df[column],df1[column],df2[column],df3[column]])
    
    plt.subplot(len(cleaning_list), 4, value+1)
    plt.bar(['Uncleaned Data Frame', 'Cleaned by Mean', 'Cleaned by Year', 'Interpolation'],df.mean())
    plt.title('Plot of {}'.format(column))

plt.tight_layout()
plt.show()
```
